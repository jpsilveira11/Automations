```python
# Função para processar os sintomas com o Hugging Face
def processar_sintomas(sintomas_selecionados):
    # Definindo as possíveis classes (categorias de doenças)
    classes_candidatas = [
        "Gripe",
        "Resfriado",
        "COVID-19",
        "Alergia",
        "Enxaqueca",
        # pode adicionar mais doenças seguindo a mesma estrutura dos itens acima (sugiro que adicione)
    ]

    # Montando a frase de entrada para o classificador
    frase_entrada = f"O paciente apresenta os seguintes sintomas: {', '.join(sintomas_selecionados)}."

    # Execute a classificação zero-shot
    resultados = classificador(frase_entrada, classes_candidatas)

    # Ordenando os resultados por probabilidade (do maior para o menor)
    resultados_ordenados = sorted(
        zip(resultados["labels"], resultados["scores"]),
        key=lambda item: item[1],
        reverse=True,
    )

    # Imprimindo os resultados (mostrando apenas os 3 mais prováveis)
    print("Os 3 diagnósticos mais prováveis são:")
    for label, score in resultados_ordenados[:3]: #Muda esse 3 se quiser mostrar mais resultados
        print(f"- {label}: {score*100:.2f}%")
        #fim da função (tudo tabulado é parte do código da função)
```

```python
# Obtendo os sintomas selecionados
def obter_sintomas_selecionados():
    sintomas_selecionados = [
        caixa_de_selecao.description
        for caixa_de_selecao in caixas_de_selecao
        if caixa_de_selecao.value
    ]
    return sintomas_selecionados
```

```python
# Função para lidar com o clique do botão
def ao_clicar_botao(b):
    sintomas_selecionados = obter_sintomas_selecionados()
    processar_sintomas(sintomas_selecionados)
```
