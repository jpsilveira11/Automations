```python
import ipywidgets as widgets
from IPython.display import display
from transformers import pipeline

# Defina os sintomas
sintomas = [
    "Febre",
    "Tosse",
    "Dor de cabeça",
    "Dor de garganta",
    "Coriza",
    "Cansaço",
    "Dor no corpo",
    "Falta de ar",
    "Diarreia",
    "Vômito",
    "Perda de paladar",
    "Perda de olfato",
]

# Crie caixas de seleção para cada sintoma
caixas_de_selecao = [
    widgets.Checkbox(value=False, description=sintoma) for sintoma in sintomas
]

# Exiba as caixas de seleção
display(widgets.VBox(caixas_de_selecao))

# Função para obter os sintomas selecionados
def obter_sintomas_selecionados():
    sintomas_selecionados = [
        caixa_de_selecao.description
        for caixa_de_selecao in caixas_de_selecao
        if caixa_de_selecao.value
    ]
    return sintomas_selecionados

# Inicialize o pipeline do Hugging Face para classificação zero-shot
classificador = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Função para processar os sintomas com o Hugging Face
def processar_sintomas(sintomas_selecionados):
    # Defina as possíveis classes (categorias de doenças)
    classes_candidatas = [
        "Gripe",
        "Resfriado",
        "COVID-19",
        "Alergia",
        "Enxaqueca",
        # adicione mais doenças aqui, se desejar
    ]

    # Crie a frase de entrada para o classificador
    frase_entrada = f"O paciente apresenta os seguintes sintomas: {', '.join(sintomas_selecionados)}."

    # Execute a classificação zero-shot
    resultados = classificador(frase_entrada, classes_candidatas)

    # Ordene os resultados por probabilidade (do maior para o menor)
    resultados_ordenados = sorted(
        zip(resultados["labels"], resultados["scores"]),
        key=lambda item: item[1],
        reverse=True,
    )

    # Imprima os resultados, mostrando apenas os 3 mais prováveis
    print("Possíveis diagnósticos (ordenados por probabilidade):")
    for label, score in resultados_ordenados[:3]:
        print(f"- {label}: {score:.4f}")

# Botão para enviar os sintomas selecionados
botao_enviar = widgets.Button(description="Enviar")
display(botao_enviar)

# Função para lidar com o clique do botão
def ao_clicar_botao(b):
    sintomas_selecionados = obter_sintomas_selecionados()
    processar_sintomas(sintomas_selecionados)

botao_enviar.on_click(ao_clicar_botao)
```
